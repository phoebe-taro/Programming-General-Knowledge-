```python
# 安裝 ColabTurtle 套件
!pip install ColabTurtle

# 匯入模組與初始化
import ColabTurtle.Turtle as Turtle
Turtle.initializeTurtle()
Turtle.speed(10)
Turtle.bgcolor("skyblue")

# 手動畫圓形的函數
def draw_circle(radius):
    for _ in range(36):  # 36 步，每步畫出小段
        Turtle.forward(2 * 3.14159 * radius / 36)  # 小步長以達到圓形效果
        Turtle.left(10)

# 畫單個雲朵
def draw_cloud(x, y):
    # 確保 x 和 y 坐標為合法值
    if x < 0:
        x = 0
    if y < 0:
        y = 0

    Turtle.penup()
    Turtle.setposition(x, y)  # 設置圓形的位置
    Turtle.pendown()
    Turtle.color("white")
    draw_circle(20)  # 畫一個圓形，模擬雲朵的一部分

# 主程式（畫出多個重疊圓形來形成雲朵）
def draw_clouds():
    # 畫出多個圓形來形成雲朵
    draw_cloud(50, 0)   # 右邊的雲
    draw_cloud(100, 20) # 右側的小雲
    draw_cloud(150, 0)  # 另一個雲

    draw_cloud(200, 20) # 下方的雲
    draw_cloud(300, 20) # 另一個雲

# 呼叫繪製雲朵的函數
draw_clouds()
```
