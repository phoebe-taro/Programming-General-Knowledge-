```python
import numpy as np
import pandas as pd
import csv
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

data1 = pd.read_csv('婚姻與生育.csv', encoding='big5')

years = data1.iloc[-20:, 0]
marriage = data1.iloc[-20:, 1]
bron = data1.iloc[-20:, 2]
divorce =  data1.iloc[-20:, 3]
```

```python
trace1 = go.Scatter(
    x = years,
    y = marriage,
    mode = 'lines+markers',
    name = '結婚對數'
)

trace2 = go.Scatter(
    x = years,
    y = bron,
    mode = 'lines+markers',
    name = '出生人數'
)

data1 = [trace1, trace2]

iplot(data1, filename='scatter-mode')
```
