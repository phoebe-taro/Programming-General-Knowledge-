```python
import numpy as np
import pandas as pd
import csv

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
data1 = pd.read_csv('gdp_taiwan.csv')
data2 = pd.read_csv('gdp_world.csv')
Taiwan = data1.loc[0][1:]
HongKong = data2.loc[94][1:]
Korea = data2.loc[124][1:]
Singapore = data2.loc[206][1:]

years = np.linspace(1960, 2017, 58)

trace1 = go.Scatter(
    x = years,
    y = Taiwan*1000000,
    mode = 'lines+markers',
    name = '臺灣'
)

trace2 = go.Scatter(
    x = years,
    y = HongKong,
    mode = 'lines+markers',
    name = '香港'
)

trace3 = go.Scatter(
    x = years,
    y = Korea,
    mode = 'lines+markers',
    name = '南韓'
)

trace4 = go.Scatter(
    x = years,
    y = Singapore,
    mode = 'lines+markers',
    name = '新加坡'
)

data = [trace1, trace2,trace3,trace4]

iplot(data, filename='scatter-mode')
```
