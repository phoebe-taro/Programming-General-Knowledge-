```python
import pandas as pd
import csv

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

data = pd.read_csv('coffee2.csv')

zone1 = data.loc[0][1:]
zone2 = data.loc[1][1:]
zone3 = data.loc[2][1:]
zone4 = data.loc[3][1:]

xName = list(data.columns.values)

trace1 = go.Bar(
    x = xName[1:5],
    y = zone1*60*1000,
    name = data.iloc[0,0]
)

trace2 = go.Bar(
    x = xName[1:5],
    y = zone2*60*1000,
    name = data.iloc[1,0]
)

trace3 = go.Bar(
    x = xName[1:5],
    y = zone3*60*1000,
    name = data.iloc[2,0]
)

trace4 = go.Bar(
    x = xName[1:5],
    y = zone4*60*1000,
    name = data.iloc[3,0]
)

data = [trace1, trace2, trace3, trace4]

iplot(data, filename='scatter-mode')
```
